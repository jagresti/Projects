```python
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
!ls "/content/drive/MyDrive/Econ148 Paper 4"
df = pd.read_stata('/content/drive/MyDrive/Econ148 Paper 4/aerdat4.dta')
df
'''

import statsmodels.api as sm
from statsmodels.formula.api import ols

def run_regression(df, outcome, cohort_filter, extra_controls=""):
    subset = df.query(cohort_filter).copy()
    if extra_controls:
        formula = f'{outcome} ~ vouch0 + djamundi + dbogota + d1993 + d1995 + d1997 + {extra_controls}'
    else:
        formula = f'{outcome} ~ vouch0 + djamundi + dbogota + d1993 + d1995 + d1997'
    model = ols(formula, data=subset).fit(cov_type='HC0')
    mean_loser = subset[subset.vouch0 == 0][outcome].mean()
    effect = model.params.get('vouch0', float('nan'))
    stderr = model.bse.get('vouch0', float('nan'))
    return mean_loser, effect, stderr, subset

# Specify cohorts
cohorts = {
    'Bogotá 1995': '(9 <= age2 <= 25) & (dbogota == 1) & (d1995 == 1)',
    'Bogotá 1997': '(9 <= age2 <= 25) & (dbogota == 1) & (d1997 == 1)',
    'Jamundí 1993': '(9 <= age2 <= 25) & (djamundi == 1) & (d1993 == 1)',
    'Combined sample': '(9 <= age2 <= 25) & (((dbogota == 1) & (d1995 == 1)) | ((dbogota == 1) & (d1997 == 1)) | ((djamundi == 1) & (d1993 == 1)))',
    'Test-takers': 'vouch0.null()'
}

# Function to create panel rows
def build_panel(panel_title, variables, base_filter="", extra_controls=""):
    results = [[panel_title] + [''] * (len(cohorts) * 2)]
    for var in variables:
        row = [var]
        for _, filt in cohorts.items():
            filt_full = f"{filt}"
            try:
                mean_loser, effect, stderr, subset = run_regression(df, var, filt_full, extra_controls)

                if var in ['age2', 'age', 'mom_sch','dad_sch', 'mom_age','dad_age']:
                    loser_group = subset[subset.vouch0 == 0][var]
                    mean_loser_se = loser_group.std()
                    loser_formatted = f"{mean_loser:.1f} ({mean_loser_se:.3f})"
                else:
                    loser_formatted = f"{mean_loser:.3f}"

                effect_se_formatted = f"{effect:.3f} ({stderr:.3f})"
                row.extend([loser_formatted, effect_se_formatted])
            except:
                row.extend(['—', '—'])
        results.append(row)

    # Add N row
    n_row = ['N']
    for _, filt in cohorts.items():
        filt_full = f"{filt}"
        try:
            subset = df.query(filt_full)
            n_loser = (subset.vouch0 == 0).sum()
            n_winner = (subset.vouch0 == 1).shape[0]
            n_row.extend([str(n_loser), str(n_winner)])
        except:
            n_row.extend(['—', '—'])
    results.append(n_row)
    return results

# Panel A
panel_a = build_panel("A. Data from PACES Application:", ['phone', 'age2', 'sex_name'])

# Panel B
cohorts = {
    'Bogotá 1995': '(9 <= age2 <= 25) & (dbogota == 1) & (d1995 == 1) & (bog95asd == 1)',
    'Bogotá 1997': '(9 <= age2 <= 25) & (dbogota == 1) & (d1997 == 1) & (bog97asd == 1)',
    'Jamundí 1993': '(9 <= age2 <= 25) & (djamundi == 1) & (d1993 == 1) & (jam93asd == 1)',
    'Combined sample': '( ((9 <= age2 <= 25) & (dbogota == 1) & (d1995 == 1) & (bog95asd == 1)) | ((9 <= age2 <= 25) & (dbogota == 1) & (d1997 == 1) & (bog97asd == 1)) | ((9 <= age2 <= 25) & (djamundi == 1) & (d1993 == 1) & (jam93asd == 1)) )',
    'Test-takers': 'vouch0.null()'
}

panel_b = build_panel("B. Data for All Attempted Contacts:", ['phone', 'age2', 'sex_name'])

# Panel C
cohorts = {
    'Bogotá 1995': '(dbogota == 1) & (d1995 == 1) & (bog95smp == 1)',
    'Bogotá 1997': '(dbogota == 1) & (d1997 == 1) & (bog97smp == 1)',
    'Jamundí 1993': '(djamundi == 1) & (d1993 == 1) & (jam93smp == 1)',
    'Combined sample': '( ((dbogota == 1) & (d1995 == 1) & (bog95smp == 1)) | ((dbogota == 1) & (d1997 == 1) & (bog97smp == 1)) | ((djamundi == 1) & (d1993 == 1) & (jam93smp == 1)) )',
    'Test-takers': '(test_tak == 1)'
}

panelC_cols = {
    'Age at time of survey': 'age',
    'Male': 'sex',
    "Mother's highest grade completed": 'mom_sch',
    "Father's highest grade completed": 'dad_sch',
    "Mother's age": 'mom_age',
    "Father's age": 'dad_age',
    "Father's wage": 'dad_mw'
}
panel_c = build_panel("C. Survey Data:", ['age','sex','mom_sch','dad_sch', 'mom_age','dad_age', 'dad_mw'])

# Combine all panels into one dataframe
full_results = panel_a + panel_b + panel_c
columns = ['Dependent variable']
for cohort in cohorts:
    columns.extend([
        f"{cohort} (Loser mean)",
        f"{cohort} (Won voucher)"
    ])

table2_full = pd.DataFrame(full_results, columns=columns)



# Relabelling
row_labels = {
    'phone': 'Has phone',
    'age2': 'Age at time of application',
    'sex_name': 'Male',
    'age': 'Age at time of survey',
    'sex': 'Male',
    'mom_sch': "Mother's highest grade completed",
    'dad_sch': "Father's highest grade completed",
    'mom_age': "Mother's age",
    'dad_age': "Father's age",
    'dad_mw': "Father's wage",
    'A. Data from PACES Application:': 'A. Data from PACES Application:',
    'B. Data for All Attempted Contacts:': 'B. Data for All Attempted Contacts:',
    'C. Survey Data:': 'C. Survey Data:',
    'N': 'N'
}

table2_full.iloc[:, 0] = table2_full.iloc[:, 0].replace(row_labels)

column_labels = {
    'Bogotá 1995': 'Bogotá 1995',
    'Bogotá 1997': 'Bogotá 1997',
    'Jamundí 1993': 'Jamundí 1993',
    'Combined sample': 'Combined sample',
    'Test-takers': 'Test-takers'
}

new_columns = ['Dependent variable']
for cohort in column_labels:
    new_columns.extend([
        f"{column_labels[cohort]} (Loser means)",
        f"{column_labels[cohort]} (Won voucher)"
    ])
table2_full.columns = new_columns

# Rearraanging
cohorts = ['Bogotá 1995', 'Bogotá 1997', 'Jamundí 1993', 'Combined sample', 'Test-takers']
subcols = ['Loser mean', 'Won voucher']
table2_full.columns = ['Dependent variable'] + [f"{c} - {s}" for c in cohorts for s in subcols]
table2_arranged = pd.DataFrame(table2_full.values, columns=pd.MultiIndex.from_tuples([('', 'Dependent variable')] + [(c, s) for c in cohorts for s in subcols]))
table2_arranged = table2_arranged.style.applymap(lambda v: 'font-style: italic' if '. ' in str(v) else '', subset=[('', 'Dependent variable')])
table2_arranged
```

```python
#Table 3 REGRESSION

# Define outcomes
outcomes = [
    'used_scholarship', 'ever_used_scholarship',
    'started_6th_private', 'started_7th_private', 'currently_private',
    'highest_grade', 'in_school',
    'finished_6th', 'finished_7th', 'finished_8th',
    'repeated_6th', 'ever_repeated', 'total_repeats', 'years_schooling'
]

# Basic and all controls
basic_controls = (
    'survey_id + interview_type + jamundi_dummy + phone_access + '
    'age + male + strata1 + strata2 + strata3 + strata4 + strata5 + strata6 + stratams + '
    'bogota_dummy + yr1993 + yr1995 + yr1997 + '
    'dmonth1 + dmonth2 + dmonth3 + dmonth4 + dmonth5 + dmonth6 + '
    'dmonth7 + dmonth8 + dmonth9 + dmonth10 + dmonth11 + dmonth12 + sex_missing'
)

all_controls = basic_controls + (
    ' + darea1 + darea2 + darea3 + darea4 + darea5 + darea6 + darea7 + darea8 + darea9 + '
    'darea10 + darea11 + darea12 + darea13 + darea14 + darea15 + darea16 + darea17 + darea18 + darea19'
)

# Regression function
def run_regression(y, controls='', sample_filter=None):
    data = Table3_1.copy()
    if sample_filter:
        data = data.query(sample_filter)
    formula = f"{y} ~ won"
    if controls:
        formula += f" + {controls}"
    model = smf.ols(formula, data=data).fit(cov_type = "HC1")
    return round(model.params['won'], 3), round(model.bse['won'], 3), int(model.nobs)

# Building table
rows = []

for y in outcomes:
    # Filters
    bog_filter = "tab3smpl == 1 and bog95smp == 1"
    if y in ['finished_7th', 'finished_8th']:
        combined_filter = "tab3smpl == 1 and (bog95smp == 1 or jam93smp == 1)"
    else:
        combined_filter = "tab3smpl == 1 and (bog95smp == 1 or bog97smp == 1 or jam93smp == 1)"

    # Loser mean
    loser_mean = round(
        Table3_1.query(f"{bog_filter} and won == 0")[y].mean(),
        3
    )

    # Bogotá ’95 regressions
    coef_nc, se_nc, n_nc = run_regression(y, controls='',            sample_filter=bog_filter)
    coef_bc, se_bc, n_bc = run_regression(y, controls=basic_controls, sample_filter=bog_filter)
    coef_all,se_all,n_all= run_regression(y, controls=all_controls,   sample_filter=bog_filter)

    # Combined‐sample regressions
    coef_cb,   se_cb,   n_cb    = run_regression(y, controls=basic_controls, sample_filter=combined_filter)
    coef_cb_all,se_cb_all,n_cb_all = run_regression(y, controls=all_controls, sample_filter=combined_filter)

    rows.append({
      'Outcome':        y,
      'Loser Mean':     loser_mean,
      'Bogotá No Ctrl': f"{coef_nc} ({se_nc})",
      'Bogotá Basic':   f"{coef_bc} ({se_bc})",
      'Bogotá +19':     f"{coef_all} ({se_all})",
      'Combined Basic': f"{coef_cb} ({se_cb})",
      'Combined +19':   f"{coef_cb_all} ({se_cb_all})",
      'Sample Size':    n_cb
    })

table3_df = pd.DataFrame(rows)
table3_df
```

```python
# REPLICATING TABLE 4

df4 = Table3_1.rename(columns={
    'won':            'vouch0',    # voucher dummy
    'survey_id':      'svy',
    'interview_type': 'hsvisit',
    'jamundi_dummy':  'djamundi',
    'phone_access':   'phone',
    'age':            'age2',
    'male':           'sex2',
    'bogota_dummy':   'dbogota',
    'yr1993':         'd1993',
    'yr1995':         'd1995',
    'yr1997':         'd1997',
    'sex_missing':    'sex_miss'
})


# 2) Regression helper
def run_reg4(y, controls='', sample_filter=None):
    data = df4.copy()
    if sample_filter:
        data = data.query(sample_filter)
    formula = f"{y} ~ vouch0"
    if controls:
        formula += " + " + controls
    m = smf.ols(formula, data=data).fit(cov_type='HC0')
    return round(m.params['vouch0'], 3), round(m.bse['vouch0'], 3), int(m.nobs)


# SAS basic‐controls
strata = [f"strata{i}" for i in range(1,7)]
months = [f"dmonth{i}" for i in range(1,13)]
basic_controls4 = " + ".join([
    'svy',
    'hsvisit',
    'djamundi',
    'phone',
    'age2',
    'sex2',
    *strata,
    'stratams',
    'dbogota',
    'd1993', 'd1995', 'd1997',
    *months,
    'sex_miss'])

# 4) Outcomes to loop over
outcomes4 = [
    'started_6th_private', 'started_7th_private',
    'currently_private', 'highest_grade',
    'in_school', 'finished_6th', 'finished_7th',
    'finished_8th', 'repeated_6th',
    'ever_repeated', 'total_repeats', 'years_schooling'
]

# 5) Build table
rows4 = []
for y in outcomes4:
    # Loser means
    male_loser = round(
        df4.query("tab3smpl == 1 and bog95smp == 1 and sex2 == 1 and vouch0 == 0")[y].mean(),
        3
    )
    female_loser = round(
        df4.query("tab3smpl == 1 and bog95smp == 1 and sex2 == 0 and vouch0 == 0")[y].mean(),
        3
    )

    # Bogotá ’95 regressions by sex2
    bog95_m = "tab3smpl == 1 and bog95smp == 1 and sex2 == 1"
    bog95_f = "tab3smpl == 1 and bog95smp == 1 and sex2 == 0"
    m95_coef, m95_se, _ = run_reg4(y, basic_controls4, bog95_m)
    f95_coef, f95_se, _ = run_reg4(y, basic_controls4, bog95_f)

    # Combined sample: drop bog97 for finish7 / finish8 only —
    if y in ['finished_7th', 'finished_8th']:
        comb_m = "tab3smpl == 1 and (bog95smp == 1 or jam93smp == 1) and sex2 == 1"
        comb_f = "tab3smpl == 1 and (bog95smp == 1 or jam93smp == 1) and sex2 == 0"
    else:
        comb_m = (
            "tab3smpl == 1 and "
            "(bog95smp == 1 or bog97smp == 1 or jam93smp == 1) and sex2 == 1"
        )
        comb_f = (
            "tab3smpl == 1 and "
            "(bog95smp == 1 or bog97smp == 1 or jam93smp == 1) and sex2 == 0"
        )
    mC_coef, mC_se, _ = run_reg4(y, basic_controls4, comb_m)
    fC_coef, fC_se, _ = run_reg4(y, basic_controls4, comb_f)

    rows4.append({
        'Outcome':           y.replace('_',' ').title(),
        'Bog95 Male':        f"{m95_coef:.3f} ({m95_se:.3f})",
        'Bog95 Female':      f"{f95_coef:.3f} ({f95_se:.3f})",
        'Combined Male':     f"{mC_coef:.3f} ({mC_se:.3f})",
        'Combined Female':   f"{fC_coef:.3f} ({fC_se:.3f})",
        'Loser Mean Male':   f"{male_loser:.3f}",
        'Loser Mean Female': f"{female_loser:.3f}"
    })


table4_df = pd.DataFrame(rows4)
table4_df
```

```python
# Turn test scores into standard deviation units

listed_variables = ['totalpts', 'math', 'reading', 'writing']
for variable in listed_variables:
    mean = np.mean(t5copy[variable])
    std = np.std(t5copy[variable])
    t5copy[variable] = (t5copy[variable] - mean) / std
```

```python
# All Applicants Panel

# Total Points
X = t5copy[['vouch0', 'tsite1', 'tsite2', 'tsite3']]
X = sm.add_constant(X)
y = t5copy['totalpts']
model_total = sm.OLS(y, X).fit()
print(model_total.summary())

# Math model
y_math = t5copy['math']
model_math = sm.OLS(y_math, X).fit()
print(model_math.summary())

# Reading model
y_reading = t5copy['reading']
model_reading = sm.OLS(y_reading, X).fit()
print(model_reading.summary())

# Writing model
y_writing = t5copy['writing']
model_writing = sm.OLS(y_writing, X).fit()
print(model_writing.summary())

# All Applicants OLS results with covariates
covariates = ['svy', 'hsvisit', 'age', 'sex', 'mom_sch', 'dad_sch',
             'mom_miss', 'dad_miss']
t5copy = t5copy[t5copy[covariates].notna().all(axis=1)]
X_full = t5copy[['vouch0', 'tsite1', 'tsite2', 'tsite3'] + covariates]
X_full = sm.add_constant(X_full)

# Total points model with covariates
y_total_co = t5copy['totalpts']
model_total_co = sm.OLS(y_total_co, X_full).fit()
print(model_total_co.summary())

# Math model with covariates
y_math_co = t5copy['math']
model_math_co = sm.OLS(y_math_co, X_full).fit()
print(model_math_co.summary())

# Reading model with covariates
y_reading_co = t5copy['reading']
model_read_co = sm.OLS(y_reading_co, X_full).fit()
print(model_read_co.summary())

# Writing model with covariates
y_writing_co = t5copy['writing']
model_writing_co = sm.OLS(y_writing_co, X_full).fit()
print(model_writing_co.summary())
```

```python
# All Applicants Pooled/Math & Reading Scores

t5copy_mixed = t5copy.dropna(subset=['math', 'reading', 'writing'])

math_data = t5copy_mixed.copy()
math_data['score'] = math_data['math']
math_data['subject'] = 1
math_data = math_data.drop(['math', 'reading', 'writing', 'totalpts'], axis = 1)

reading_data = t5copy_mixed.copy()
reading_data['score'] = reading_data['reading']
reading_data['subject'] = 2
reading_data = reading_data.drop(['math', 'reading', 'writing', 'totalpts'], axis = 1)

writing_data = t5copy_mixed.copy()
writing_data['score'] = writing_data['writing']
writing_data['subject'] = 3
writing_data = writing_data.drop(['math', 'reading', 'writing', 'totalpts'], axis = 1)

pooled_data = pd.concat([math_data, reading_data, writing_data], ignore_index = True)

print("Table 3 Column 3 - Math and Reading")
math_reading = pooled_data[pooled_data['subject'] != 3]
math_and_reading = smf.mixedlm("score ~ vouch0 + tsite1 + tsite2", data = math_reading, groups = math_reading['id'])
m_r_final = math_and_reading.fit()
print(m_r_final.summary())

# Math & Reading with covariates
m_r_co = smf.mixedlm("score ~ vouch0 + tsite1 + tsite2 + svy + hsvisit + age + sex + mom_sch + dad_sch + mom_miss + dad_miss",
                 data = math_reading,
                 groups = math_reading['id'])
math_reading_co = m_r_co.fit()
print(math_reading_co.summary())

# Pooled Model
print("Table 3 Column 3 - Pooled")
pooled = smf.mixedlm("score ~ vouch0 + tsite1 + tsite2",
                  data = pooled_data,
                  groups = pooled_data['id'])
pooled_fit = pooled.fit()
print(pooled_fit.summary())

# Pooled model with covariates
print("Table 3 Column 4 - Pooled with Covariates")
pooled_co = smf.mixedlm("score ~ vouch0 + tsite1 + tsite2 + svy + hsvisit + age + sex + mom_sch + dad_sch + mom_miss + dad_miss",
                 data = pooled_data,
                 groups = pooled_data['id'])
pooled_co_fit = pooled_co.fit()
print(pooled_co_fit.summary())
```
